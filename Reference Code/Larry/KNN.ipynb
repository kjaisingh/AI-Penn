## Supervised Learning vs Unsupervised Learning

Supervised learning is the most common.  Its observed data is associated with a label (either observed or specified by experts).  Once learned, the model will be used to predict label on new unseen data.
<br><br>


Unsupervised learning involves modeling the features of a dataset without reference to any label.  Hence it is often described as "letting the dataset speak for itself." These models include structurable tasks such as clustering (identifying distinct groups of data).



## Supervised Learning: Classification vs. Regression

When you are predicting a type, a category or a label, you are performing **Classification** task.
<br>
When you are predicting a numerical value or a continuous quantity, you are typically performing **Regression** task.
<br><br>
For example, predicting stock price is a **Regression** task, while predicting the winner of a presidential election is **Classification** task.

## Iris Flower

The Iris flower data set, or Fisher's Iris data set, is a multivariate data set introduced by the British biologist Ronald Fisher in his 1936 paper.  The data is collected to quantify the morphologic variation of Iris flowers of three related species, Iris Versicolor, Iris Virginica, and Iris Setosa. 
<br><br>
Four features were measured from each sample: the length and the width of the sepals and petals, in centimeters. 
<br><br>
Two of the three species were collected in the Gaspé Peninsula (Quebec, Canada) "all from the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus".

# KNN Machine Learning Model

Welcome to our first machine learning model!  This is one of the simplest ML models and the most intuitive too.

## KNN (K-Nearest Neighbors) Model

KNN algorithms <u>classify new data points based on a similarity measures (e.g. distance function)</u>.  **Classification is done by a majority vote to its neighbors**:  the label is assigned to t**he class which has the most nearest neighbors**.

![alt text](http://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1531424125/KNN_final1_ibdm8a.png)

## KNN using Scikit-learn Machine Learning Library

Very popular machine learning library in Python.  Provide many machine learning models in easy-to-use ways.

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import  accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

#### Split data into features and labels 

```
features = df[['sepal length in cm', 
               'sepal width in cm', 
               'petal length in cm', 
               'petal width in cm']]
labels = df['class']
```

```
type(features)
```

```
type(labels)
```

```
labels.head()
```

#### Convert class labels into numbers

Class labels are strings.  We need to convert them into numbers that could be understood by computers.

```
encoder = LabelEncoder()
y = encoder.fit_transform(labels)
```

As you can see Iris-setosa Iris-versicolor Iris-virginica are converted to 0, 1, 2 respectively.

```
y
```

#### Generate training set and test set

It is common practice when performing a (supervised) machine learning training to hold out part of the available data as a test set.  sk-learn will help us quickly sample a training set while holding out certain percentage of data for testing (evaluating) our model.  The command below will have a 7-3 split of training data vs. testing data.

```
# x_train, y_train will be used for training the model
# while x_test, y_test will be used for testing/evaluating the trained model
# random_state is the seed used by random number generator
# stratify=y ensures data is split in the stratified fashion
x_train, x_test, y_train, y_test = train_test_split(features, y, test_size = 0.3, 
                                                    random_state = 101, stratify=y)
```

The data count should show stratified effect.

```
print('Labels counts in y:', np.bincount(y))
print('Labels counts in y_train:', np.bincount(y_train))
print('Labels counts in y_test:', np.bincount(y_test))
```

#### Feature Scaling

Feature scaling is a method used to standardize the range of independent variables or features of data. In data processing, it is also known as data normalization and is generally performed during the data preprocessing step.

```
x_train.head()
```

```
sc = StandardScaler()

sc.fit(features)
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)

type(x_train_std)
```

```
x_train_std[:5]
```

#### Create KNN model

For parameter p below.  When p = 1, this is equivalent to using manhattan_distance, and euclidean_distance for p = 2 (default).
![alt text](https://www.bogotobogo.com/python/scikit-learn/images/k-NN/Euclidean-vs-Manhattan.png)

```
# K=3 below
knn_model = KNeighborsClassifier(n_neighbors=7, p=2)
```

#### Run KNN Machine Learning Process (Train model, Fit data)

```
knn_model.fit(x_train_std, y_train);
```

#### Run Prediction on test set

```
knn_predict = knn_model.predict(x_test_std)
```

#### Is the prediction accurate?

```
print('knn - test accuracy ', accuracy_score(knn_predict, y_test))
```

Is the prediction on train set 100% accurate?

```
y_train_predict = knn_model.predict(x_train_std)
print('knn - train accuracy ', accuracy_score(y_train_predict, y_train))
```

####Identify the test samples predicted wrong

Get the index of the test samples predicted wrong.

```
test_predict_wrong = np.nonzero(knn_predict-y_test)
test_predict_wrong
```

```
knn_predict[test_predict_wrong]
```

```
y_test[test_predict_wrong]
```

Retrieve the normalized test data:

```
x_train_std[test_predict_wrong]
```

## Visualize Decision/Plot Decision Regions

It is hard to visualize Machine Learning, which is often called as **Black Box**.  We will try to visualize the result below.

### Helper functions.  Just run and skip.

```
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):

    # setup marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])

    # plot the decision surface
    x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())

    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], 
                    y=X[y == cl, 1],
                    alpha=0.8, 
                    c=colors[idx],
                    marker=markers[idx], 
                    label=cl, 
                    edgecolor='black')

    # highlight test samples
    if test_idx:
        # plot all samples
        X_test, y_test = X[test_idx, :], y[test_idx]
        c_test = [colors[i] for i in y_test]

        plt.scatter(X_test[:, 0],
                    X_test[:, 1],
                    c=c_test,
                    edgecolor='black',
                    alpha=0.8,
                    linewidth=1,
                    marker='o',
                    s=100, 
                    label='test set')
```

### Pick 2 features to visualize Decision Boundary

We will pick features: Petal Length and Petal Width.

```
X_train_std = x_train_std[:, 2:]
X_test_std  = x_test_std[:, 2:]

# Fit KNN model again
knn_model.fit(X_train_std, y_train);
```

Now visualized the Decision Boundary.

```
X_combined = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))

plot_decision_regions(X_combined, y_combined, 
                      classifier=knn_model, test_idx=range(105, 150))

plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
```

## KNN model: Pros and Cons

###Pros

1.   It is extremely easy to implement
2.   It is, so called, lazy learning algorithm and therefore requires no training prior to making real time predictions. This makes the KNN algorithm much faster than other algorithms that require training.
3.   Since the algorithm requires no training before making predictions, new data can be added seamlessly.
4. It's easy to understand.

### Cons



1.   The KNN algorithm doesn't work well with high dimensional data because with large number of dimensions, it becomes difficult for the algorithm to calculate distance in each dimension.
2.   The KNN algorithm has a high prediction cost for large datasets. This is because in large datasets the cost of calculating distance between new point and each existing point becomes higher.
3. Finally, the KNN algorithm doesn't work well with categorical features since it is difficult to find the distance between dimensions with categorical features.
